```python
sns.countplot(data['NumberOfTime30-59DaysPastDueNotWorse'], hue= data['SeriousDlqin2yrs'])
data['NumberOfTime30-59DaysPastDueNotWorse'].value_counts()
```

```python
def number_of_duenotoworse(x):
    if x == 0:
        return '0'
    elif x == 1:
        return '1'
    elif x == 2:
        return '2'
    elif x ==3:
        return '3'
    elif x== 4:
        return '4'
    elif x == 5:
        return '5'
    else: 
        return '>6'
    
number_of3059grouping = data['NumberOfTime30-59DaysPastDueNotWorse'].apply(number_of_duenotoworse)
number_of6089grouping = data['NumberOfTime60-89DaysPastDueNotWorse'].apply(number_of_duenotoworse)
data.insert(6, "number_of6089grouping", number_of6089grouping, True)
data.insert(1, "number_of3059grouping", number_of3059grouping, True)
```

```python
NumberRealEstateLoansOrLinesgrouping = data['NumberRealEstateLoansOrLines'].apply(number_of_duenotoworse)
data.insert(7, "NumberRealEstateLoansOrLinesgrouping", NumberRealEstateLoansOrLinesgrouping, True)

NumberOfTimes90DaysLategrouping = data['NumberOfTimes90DaysLate'].apply(number_of_duenotoworse)
data.insert(8, "NumberOfTimes90DaysLategrouping", NumberOfTimes90DaysLategrouping, True)
```
