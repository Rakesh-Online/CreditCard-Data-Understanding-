```python
sns.countplot(data['NumberOfTime30-59DaysPastDueNotWorse'], hue= data['SeriousDlqin2yrs'])
data['NumberOfTime30-59DaysPastDueNotWorse'].value_counts()
```

```python
def number_of3059duenotoworse(x):
    if x == 0:
        return '0'
    elif x == 1:
        return '1'
    elif x == 2:
        return '2'
    elif x ==3:
        return '3'
    elif x== 4:
        return '4'
    elif x == 5:
        return '5'
    else: 
        return '>6'
    
number_of3059grouping = data['NumberOfTime30-59DaysPastDueNotWorse'].apply(number_of3059duenotoworse)
data.insert(1, "number_of3059grouping", number_of3059grouping, True)
```

```python
sns.countplot(data['number_of3059grouping'], order= ['0','1','2','3','4','5', '>6'], hue = data['SeriousDlqin2yrs'])
data['number_of3059grouping'].value_counts()
```
